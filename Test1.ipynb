```python
# view_phase1.py
import pickle, gzip
import pandas as pd

PHASE1_FILE = "phase1_trajectories.pkl.gz"

with gzip.open(PHASE1_FILE, "rb") as f:
    payload = pickle.load(f)

# Check keys (usually 'data', maybe others)
print("Available keys in payload:", list(payload.keys()))

trajectories = payload["data"]
df = pd.DataFrame(trajectories)

print(f"✅ Loaded {len(df)} trajectories")
print(df.head())

# Optional: save sample to CSV
df.head(100).to_csv("sample_phase1.csv", index=False)
```

```python
# view_phase2.py
import pickle, gzip
import pandas as pd

PHASE2_FILE = "phase2_clusters.pkl.gz"

with gzip.open(PHASE2_FILE, "rb") as f:
    clusters = pickle.load(f)

print(f"✅ Loaded {len(clusters)} clusters")

# view_phase2.py  (updated snippet)
df = pd.DataFrame([
    {
        "cluster_id":  c["cluster_id"],
        "rep_subpath": c["rep_subpath"],
        "gamma":       c["gamma"],
        "n_members":   len(c["members"]),
        "trip_ids":    [m["trip_id"] for m in c["members"]]   # NEW
    }
    for c in clusters
])

print(df.head())


# Optional: save full summary
df.to_csv("summary_phase2_clusters.csv", index=False)

# Save full cluster members (if needed)
# with open("phase2_full_debug.json", "w") as f:
#     json.dump(clusters, f, indent=2)
```
